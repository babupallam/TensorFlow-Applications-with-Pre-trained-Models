```python
from mtcnn import MTCNN
import cv2

# Initialize the MTCNN detector
detector = MTCNN()
```

```python

# Load the image from file, handling potential errors
try:
    img = cv2.imread('objDetn1.jpg')
    if img is None:
        raise FileNotFoundError("Image not found at 'objDetn1.jpg'")
except FileNotFoundError as e:
    print(e)
    # Handle the case where the image is not found, e.g., exit the script or use a default image
else:
    # Convert the image from BGR to RGB
    rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    # Continue with your image processing tasks
```

```python

# Detect faces in the image
faces = detector.detect_faces(rgb_img)
```

```python

# Draw bounding boxes around detected faces
for face in faces:
    x, y, w, h = face['box']
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
    # Optionally, you can add a label or other annotations here
    cv2.putText(img, 'Face', (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
```
